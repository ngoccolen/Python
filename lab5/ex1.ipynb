```python
import numpy as np
```

```python
a = np.array([1, 2, 3])   #Tạo mảng 1 chiều với các phần tử là 1,2,3
```

```python
print(type(a)) #In ra kiểu dữ liệu của a
```

```python
print(a.shape)  #In ra kích thước của mảng
```

```python
print(a[0], a[1], a[2]) #In ra phần tử thứ 1, 2, 3 của mảng
```

```python
a[0] = 5 #Thay đổi phần tử đầu tiên của mảng từ 1 thành 5.
```

```python
print(a) #In a
```

```python
b = np.array([[1,2,3],[4,5,6]]) #Tạo mảng 2 chiều
```

```python
print(b.shape) #In ra kích thước của mảng b
```

```python
print(b[0, 0], b[0, 1], b[1, 0]) #In ra phần tử ở vị trí b[hàng,cột]
```
