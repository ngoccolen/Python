```python
import numpy as np
```

```python
x = np.array([[1,2],[3,4]]) #Tạo ma trận 2x2
```

```python
y = np.array([[5,6],[7,8]]) #Tạo ma trận 2x2
```

```python
v = np.array([9,10]) #Tạo ma trận 2x2
```

```python
w = np.array([11, 12]) #Tạo ma trận 2x2
```

```python
print(v.dot(w)) #Tính tích vô hướng giữa v và w
```

```python
print(np.dot(v, w)) #Tính tích vô hướng giữa v và w
```

```python
print(x.dot(v)) #Tính tích vô hướng giữa x và v
```

```python
print(np.dot(x, v)) #Tính tích vô hướng giữa x và v
```

```python
print(x.dot(y)) #Tính tích vô hướng giữa x và y
```

```python
print(np.dot(x, y)) #Tính tích vô hướng giữa x và y
```
