```python
import numpy as np
```

```python
a = np.array([[1,2], [3, 4], [5, 6]]) #Tạo mảng a với kích thước 3x2
```

```python
bool_idx = (a > 2) #So sánh từng phần tử của a với 2
```

```python
print(bool_idx) #In ra True nếu lớn hơn 2 và False nếu nhỏ hơn 2
```

```python
print(a[bool_idx]) #In ra mảng boolean
```

```python
print(a[a > 2]) #Lọc và trả về các giá trị trong a tại vị trí True trong bool_idx
```
