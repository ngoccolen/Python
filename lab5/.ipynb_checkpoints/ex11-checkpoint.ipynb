```python
import numpy as np
```

```python
x = np.array([[1,2],[3,4]]) #Tạo mảng có kích thước 2x2
```

```python
print(np.sum(x)) #Cộng tất cả phần tử của x
```

```python
print(np.sum(x, axis=0)) #Cộng các phần tử ở cột 1
```

```python
print(np.sum(x, axis=1)) #Cộng các phần tử ở cột 2
```

```python
x = np.array([[1,2], [3,4]]) 
```

```python
print(x) #In ra mảng x
```

```python
print(x.T) #Ma trận chuyển vị
```

```python
v = np.array([1,2,3]) #Tạo mảng có kích thước 1x3
```

```python
print(v) #In ra mảng v
```

```python
print(v.T) #Chuyển vị
```
