```python
import numpy as np
```

```python
x = np.array([1, 2]) #Tạo mảng x
```

```python
print(x.dtype) #In ra kiểu dữ liệu của mảng x
```

```python
x = np.array([1.0, 2.0]) #Tạo mảng x gồm các phần tử là 1.0 2.0
```

```python
print(x.dtype) #In ra kiểu dữ liệu của mảng x
```

```python
x = np.array([1, 2], dtype=np.int64) #Tạo mảng số nguyên nhưng ép kiểu dữ liệu thành int64
```

```python
print(x.dtype) #In ra kiểu dữ liệu của mảng x
```
