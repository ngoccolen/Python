```python
import numpy as np
```

```python
a = np.zeros((2,2))   #Tạo mảng 2 chiều kích thước (2,2) với các phần tử là 0
```

```python
b = np.ones((1,2)) #Tạo mảng 2 chiều kích thước (1,2) với các phần tử là 1
```

```python
c = np.full((2,2), 7) #Tạo mảng 2 chiều kích thước (2,2) và điền tất cả phần tử là 7
```

```python
d = np.eye(2) #Tạo mảng 2 chiều với đường chéo chính là 1, còn lại là 0
```

```python
e = np.random.random((2,2)) #Tạo một mảng 2 chiều kích thước (2, 2) với các giá trị ngẫu nhiên từ phân phối đều trong khoảng [0, 1).
```
