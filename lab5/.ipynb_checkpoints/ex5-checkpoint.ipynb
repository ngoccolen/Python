```python
import numpy as np
```

```python
a = np.array([[1,2], [3, 4], [5, 6]]) #Tạo mảng a dạng 2D
```

```python
print(a[[0, 1, 2], [0, 1, 0]]) #Lấy theo từng cặp chỉ số
```

```python
print(np.array([a[0, 0], a[1, 1], a[2, 0]])) #Lấy a[0,0] = 1 a[1,1] = 4 a[2,0] = 5
```

```python
print(a[[0, 0], [1, 1]]) #In ra giá trị a[0,1] → 2 a[0,1] → 2
```

```python
print(np.array([a[0, 1], a[0, 1]])) #Lấy hai lần a[0,1] → 2
```
