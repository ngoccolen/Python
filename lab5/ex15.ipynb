```python
import numpy as np
```

```python
v = np.array([1,2,3]) #v có 3 phần tử (shape = (3,))
```

```python
w = np.array([4,5]) #w có 2 phần tử (shape = (2,))
```

```python
print(np.reshape(v, (3, 1)) * w) #biến v thành vector cột và nhân với w
```

```python
x = np.array([[1,2,3], [4,5,6]]) #Tạo ma trận 2x3
```

```python
print(x + v) #cộng từng hàng của x với v
```

```python
print((x.T + w).T) #Cộng x.T với w. Sau đó .T lần nữa để trở lại shape ban đầu 2×3
```

```python
print(x + np.reshape(w, (2, 1))) #biến w thành vector cột 2×1, cộng từng cột của x với các giá trị 4 và 5
```
