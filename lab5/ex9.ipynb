```python
import numpy as np
```

```python
x = np.array([[1,2],[3,4]], dtype=np.float64) #Tạo ma trận 2×2 kiểu float64
```

```python
y = np.array([[5,6],[7,8]], dtype=np.float64) #Tạo ma trận 2×2 kiểu float64
```

```python
print(x + y) #Cộng từng phần tử của hai ma trận.
```

```python
print(np.add(x, y)) #Cộng từng phần tử của hai ma trận.
```

```python
print(x - y) #Trừ từng phần tử
```

```python
print(np.subtract(x, y)) #Trừ từng phần tử
```

```python
print(x * y) #Nhân từng phần tử
```

```python
print(np.multiply(x, y)) #Nhân từng phần tử
```

```python
print(x / y) #Chia từng phần tử.
```

```python
print(np.divide(x, y)) #Chia từng phần tử
```

```python
print(np.sqrt(x)) #Lấy căn bậc hai từng phần tử trong x.
```
