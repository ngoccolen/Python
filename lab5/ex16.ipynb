```python
import numpy as np
```

```python
v = np.array([1,2,3]) #Tạo vector v có 3 phần tử.
```

```python
w = np.array([4,5]) #Tạo vector w có 2 phần tử
```

```python
print(np.reshape(v, (3, 1)) * w) #biến v thành vector cột 3×1 rồi nhân với w
```

```python
x = np.array([[1,2,3], [4,5,6]]) #Ma trận x kích thước 2×3
```

```python
print(x + v) #cộng từng hàng của x với v
```

```python
print((x.T + w).T) #x.T là ma trận 3×2. Cộng vector w (2 phần tử) theo cột của x.T, .T lại để quay về kích thước 2×3.
```

```python
print(x + np.reshape(w, (2, 1))) #biến w thành vector cột 2×1
```

```python
print(x * 2) #Nhân từng phần tử của x với 2
```
